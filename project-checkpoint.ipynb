4.2. Complete Logistic Regression Implementation (12 points)
Develop a comprehensive `LogisticRegression` class that implements the algorithm entirely from scratch using only NumPy for numerical computations. Your implementation must include:

- Constructor: `__init__(self, learning_rate=0.01, max_iterations=1000, tolerance=1e-6)` with parameters for controlling the optimization process
- Sigmoid function: `sigmoid(self, z)` that handles numerical stability for extreme values
- Cost computation:`compute_cost(self, y_true, y_pred_proba)` that calculates the cross-entropy loss
- Gradient calculation:`compute_gradients(self, X, y, y_pred_proba)` that derives partial derivatives
- Training method:`fit(self, X, y)` that implements gradient descent with convergence monitoring
- Probability prediction: `predict_proba(self, X)` returning class probabilities
- Binary prediction: `predict(self, X)` returning discrete class labels
- Training history: Store cost values and parameter updates for analysis

Ensure your implementation handles edge cases such as numerical overflow in the exponential function and implements proper convergence criteria.

```python

import numpy as np

class LogisticRegression:
    """
    Logistic Regression implementation from scratch using Gradient Descent.
    The model learns a linear decision boundary by optimizing the 
    cross-entropy (log-loss) cost function.
    """

    def __init__(self, learning_rate=0.01, max_iterations=1000, tolerance=1e-6):
        """
        Initialize model hyperparameters and placeholders for parameters.

        Parameters:
        - learning_rate: step size for gradient descent updates.
        - max_iterations: maximum number of iterations (epochs) for training.
        - tolerance: stopping threshold for convergence (based on cost change).
        """
        self.learning_rate = learning_rate
        self.max_iterations = max_iterations
        self.tolerance = tolerance

        # Model parameters (weights & bias)
        self.weights = None
        self.bias = None

        # To track the training process over time
        self.cost_history = []    # Cost function values per iteration
        self.weight_history = []  # Weight values per iteration (for visualization)

    def sigmoid(self, z):
        """
        Compute the sigmoid (logistic) function in a numerically stable way.
        This maps any real-valued number into the range (0, 1),
        which can be interpreted as a probability.

        Formula: œÉ(z) = 1 / (1 + e^(-z))
        """
        # Clip values to avoid overflow in exp(-z)
        z = np.clip(z, -500, 500)
        return 1 / (1 + np.exp(-z))

    def compute_cost(self, y_true, y_pred_proba):
        """
        Compute the cross-entropy loss (log-loss), which measures 
        the difference between predicted probabilities and true labels.

        Parameters:
        - y_true: actual binary labels (0 or 1)
        - y_pred_proba: predicted probabilities from the sigmoid function

        Returns:
        - cost: scalar value representing the average loss over all samples
        """
        m = len(y_true)
        epsilon = 1e-10  # Small constant to avoid log(0)
        cost = -(1/m) * np.sum(
            y_true * np.log(y_pred_proba + epsilon) +
            (1 - y_true) * np.log(1 - y_pred_proba + epsilon)
        )
        return cost

    def compute_gradients(self, X, y, y_pred_proba):
        """
        Compute the gradients of the cost function with respect to
        the model parameters (weights and bias).

        Gradients indicate the direction and rate of change of the cost function.

        Returns:
        - dw: gradient of the weights
        - db: gradient of the bias
        """
        m = X.shape[0]  # number of samples
        dw = (1/m) * np.dot(X.T, (y_pred_proba - y))
        db = (1/m) * np.sum(y_pred_proba - y)
        return dw, db

    def fit(self, X, y):
        """
        Train the logistic regression model using gradient descent.

        Steps:
        1. Initialize weights and bias to zero.
        2. Compute predictions using the sigmoid function.
        3. Calculate the cost (cross-entropy loss).
        4. Compute gradients for weights and bias.
        5. Update parameters in the opposite direction of the gradients.
        6. Repeat until convergence or maximum iterations reached.
        """
        m, n = X.shape  # m: number of samples, n: number of features
        self.weights = np.zeros((n, 1))  # Initialize weights to zero
        self.bias = 0                    # Initialize bias to zero
        y = y.reshape(-1, 1)             # Ensure y is a column vector

        for i in range(self.max_iterations):
            # ---- Forward propagation ----
            z = np.dot(X, self.weights) + self.bias      # Linear combination
            y_pred_proba = self.sigmoid(z)               # Apply sigmoid to get probabilities

            # ---- Compute cost ----
            cost = self.compute_cost(y, y_pred_proba)
            self.cost_history.append(cost)

            # ---- Compute gradients ----
            dw, db = self.compute_gradients(X, y, y_pred_proba)
            self.weight_history.append(self.weights.copy())

            # ---- Parameter update (Gradient Descent step) ----
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

            # ---- Check for convergence ----
            # Stop if cost improvement is smaller than tolerance
            if i > 0 and abs(self.cost_history[-2] - cost) < self.tolerance:
                print(f"Converged after {i} iterations.")
                break

    def predict_proba(self, X):
        """
        Predict class probabilities for input data X.

        Returns:
        - probabilities between 0 and 1 for the positive class (y=1)
        """
        z = np.dot(X, self.weights) + self.bias
        return self.sigmoid(z)

    def predict(self, X):
        """
        Predict final class labels (0 or 1) based on a 0.5 threshold.

        Returns:
        - Binary predictions (0 or 1)
        """
        return (self.predict_proba(X) >= 0.5).astype(int)
```

```python
# =============================================
# 4.5: MODEL EVALUATION METRICS AND PLOTS
# =============================================

# Import essential performance metrics and plotting tools from sklearn
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, roc_curve, roc_auc_score
)

def evaluate_model_full(y_true, y_pred, y_proba):
    """
    Compute a comprehensive set of evaluation metrics for binary classification.

    Parameters:
    - y_true: true class labels (0 or 1)
    - y_pred: predicted class labels from the model
    - y_proba: predicted probabilities for the positive class (y=1)

    Returns:
    - Dictionary containing accuracy, precision, recall, specificity, F1-score,
      AUC, and the confusion matrix.
    """

    # --- Core classification metrics ---
    acc = accuracy_score(y_true, y_pred)                   # Overall correctness of predictions
    prec = precision_score(y_true, y_pred, zero_division=0) # Of all predicted positives, how many are correct
    rec = recall_score(y_true, y_pred, zero_division=0)     # Of all actual positives, how many were detected
    f1 = f1_score(y_true, y_pred, zero_division=0)          # Harmonic mean of precision and recall (balances both)

    # --- Confusion Matrix and derived metric (Specificity) ---
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()                             # Extract matrix components
    # Specificity: measures how well the model identifies negatives
    specificity = tn / (tn + fp) if (tn + fp) > 0 else 0.0

    # --- Area Under the ROC Curve (AUC) ---
    auc = roc_auc_score(y_true, y_proba)                    # Measures overall discriminative ability

    # Combine all metrics into a dictionary for easy access and reporting
    metrics = {
        "accuracy": acc,
        "precision": prec,
        "recall": rec,
        "specificity": specificity,
        "f1": f1,
        "auc": auc,
        "confusion_matrix": cm
    }
    return metrics


def print_metrics(metrics, label="Model"):
    """
    Display evaluation results in a clear and structured format.
    
    Parameters:
    - metrics: dictionary output from evaluate_model_full()
    - label: optional string to name the model being evaluated
    """
    print(f"--- {label} Metrics ---")
    print(f"Accuracy    : {metrics['accuracy']:.4f}")
    print(f"Precision   : {metrics['precision']:.4f}")
    print(f"Recall      : {metrics['recall']:.4f}")
    print(f"Specificity : {metrics['specificity']:.4f}")
    print(f"F1-score    : {metrics['f1']:.4f}")
    print(f"AUC         : {metrics['auc']:.4f}")
    print("\nConfusion Matrix:")
    print(metrics['confusion_matrix'])


def plot_confusion(cm):
    """
    Plot a heatmap representation of the confusion matrix.
    Helps visualize how many samples were correctly or incorrectly classified.
    
    Parameters:
    - cm: 2x2 confusion matrix
    """
    plt.figure(figsize=(4,3))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title("Confusion Matrix")
    plt.show()

    # Interpretation:
    # - Diagonal cells (top-left and bottom-right) = correct predictions
    # - Off-diagonal cells = misclassifications


def plot_roc(y_true, y_proba, title="ROC Curve"):
    """
    Plot the Receiver Operating Characteristic (ROC) curve
    and compute the AUC (Area Under Curve).
    
    The ROC curve shows the trade-off between:
    - True Positive Rate (Sensitivity)
    - False Positive Rate (1 - Specificity)
    """
    fpr, tpr, _ = roc_curve(y_true, y_proba)        # Compute FPR and TPR at various thresholds
    auc_val = roc_auc_score(y_true, y_proba)        # Compute AUC score

    plt.figure(figsize=(6,5))
    plt.plot(fpr, tpr, label=f"AUC = {auc_val:.4f}", color='darkorange', lw=2)
    plt.plot([0,1], [0,1], 'k--', alpha=0.7)        # Diagonal line = random guessing
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(title)
    plt.legend(loc="lower right")
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.show()

    # Interpretation:
    # A curve closer to the top-left corner indicates better performance.
    # AUC close to 1.0 = excellent model; near 0.5 = random guessing.


# Step 1: Generate predictions for the test set
y_test_proba = model.predict_proba(X_test)   # Predicted probabilities (used for ROC/AUC)
y_test_pred = model.predict(X_test)          # Final class predictions (0 or 1)

# Step 2: Compute all performance metrics for our custom model
metrics_custom = evaluate_model_full(y_test, y_test_pred, y_test_proba)

# Step 3: Print metrics in an organized and readable format
print_metrics(metrics_custom, label="Custom LogisticRegression")

# Step 4: Visualize performance using confusion matrix and ROC curve
plot_confusion(metrics_custom["confusion_matrix"])
plot_roc(y_test, y_test_proba, title="Custom Model ROC Curve")
```

```python

# 5.2 REGULARIZATION ANALYSIS (L1 Logistic Regression)

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score
import numpy as np
import matplotlib.pyplot as plt

# --- Define range of regularization strengths (C = 1/Œª) ---
# Smaller C => stronger regularization
C_values = np.logspace(-4, 4, 15)

# --- Storage for results ---
num_nonzero = []   # number of active (nonzero) weights
accuracy_list = [] # test accuracy
f1_list = []       # F1 score
auc_list = []      # ROC-AUC score

print(" TRAINING L1-REGULARIZED LOGISTIC REGRESSION ")

for C in C_values:
    # Initialize model with L1 penalty (encourages sparsity)
    model_l1 = LogisticRegression(
        penalty='l1',
        solver='liblinear',
        C=C,
        max_iter=1000,
        random_state=42
    )
    
    # Train on MNIST (3 vs not-3)
    model_l1.fit(X_train, y_train)
    
    # Evaluate on test set
    y_pred = model_l1.predict(X_test)
    y_proba = model_l1.predict_proba(X_test)[:, 1]
    
    # Compute metrics
    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_proba)
    
    # Count non-zero coefficients (model complexity)
    nnz = np.sum(model_l1.coef_ != 0)
    
    # Store results
    num_nonzero.append(nnz)
    accuracy_list.append(acc)
    f1_list.append(f1)
    auc_list.append(auc)
    
    print(f"C={C:>8.4f} | Nonzero Coeffs={nnz:4d} | Acc={acc:.4f} | F1={f1:.4f} | AUC={auc:.4f}")

# VISUALIZATION: Model Complexity and Accuracy

fig, ax1 = plt.subplots(figsize=(8, 5))

# Left axis: model complexity (# non-zero weights)
ax1.plot(C_values, num_nonzero, color='blue', marker='o', label='Non-zero Coefficients')
ax1.set_xscale('log')
ax1.set_xlabel("Regularization Parameter (C)")
ax1.set_ylabel("Number of Non-zero Coefficients", color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Right axis: test accuracy
ax2 = ax1.twinx()
ax2.plot(C_values, accuracy_list, color='red', marker='s', label='Test Accuracy')
ax2.set_ylabel("Test Accuracy", color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title("Model Complexity vs Accuracy (L1 Regularization)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# VISUALIZATION: F1-Score and AUC

fig, ax1 = plt.subplots(figsize=(8, 5))

# Left axis: F1-score
ax1.plot(C_values, f1_list, color='green', marker='o', label='F1-score')
ax1.set_xscale('log')
ax1.set_xlabel("Regularization Parameter (C)")
ax1.set_ylabel("F1-score", color='green')
ax1.tick_params(axis='y', labelcolor='green')

# Right axis: AUC
ax2 = ax1.twinx()
ax2.plot(C_values, auc_list, color='orange', marker='s', label='AUC-ROC')
ax2.set_ylabel("AUC", color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

plt.title("Performance Metrics vs Regularization (L1)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# ANALYSIS AND INTERPRETATION

# Identify the C value that gives the best F1-score (balanced performance)
best_index = np.argmax(f1_list)
optimal_C = C_values[best_index]
print(" REGULARIZATION ANALYSIS RESULTS ")
print(f"Optimal C value (based on F1-score): {optimal_C:.4f}")
print(f"Corresponding Metrics:")
print(f" - Accuracy : {accuracy_list[best_index]:.4f}")
print(f" - F1-score : {f1_list[best_index]:.4f}")
print(f" - AUC      : {auc_list[best_index]:.4f}")
print(f" - Active features: {num_nonzero[best_index]} / {X_train.shape[1]}")

```

Discussion:
1. Effect of Regularization Strength (C):
   - When C is small (e.g., 1e-4), strong regularization forces most coefficients
     to zero ‚Üí very simple model with high bias (underfitting).
   - As C increases, more coefficients become nonzero, model complexity rises,
     and accuracy/F1 improve ‚Äî up to an optimal point.
   - Beyond that, overly large C (weak regularization) may cause overfitting
     to training noise, slightly reducing test performance.

2. L1 Regularization and Sparsity:
   - L1 regularization applies an absolute penalty on coefficients.
     This pushes many weights exactly to zero, performing *implicit feature selection*.
   - In high-dimensional data like MNIST (784 features), this helps
     identify only the most relevant pixels contributing to ‚Äú3‚Äù.

3. Bias-Variance Tradeoff:
   - Small C ‚Üí strong regularization ‚Üí high bias, low variance (underfit).
   - Large C ‚Üí weak regularization ‚Üí low bias, high variance (overfit).
   - Optimal C provides balance ‚Äî lowest total error and best generalization.

4. Model Complexity vs Performance:
   - The first plot shows a rapid increase in active coefficients with C.
   - However, performance (accuracy, F1, AUC) stabilizes quickly,
     meaning many features add little value beyond a certain point.
   - The optimal regularization yields both compactness and good test metrics.

In summary, L1 regularization helps combat overfitting in this
high-dimensional MNIST problem by shrinking irrelevant pixel weights
to zero, improving interpretability and generalization.

Calculate key metrics: accuracy, precision, recall, F1-score, and AUC

```python

# 5.3 MODEL COMPARISON (WITH COMMENTS)
# --- Import necessary libraries ---
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, roc_curve, precision_recall_curve,
    average_precision_score
)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# STEP 1: Set best C value from 5.2
# Optimal L1 regularization strength from Section 5.2 analysis
best_C = 193.07  # determined by highest F1-score

# STEP 2: Helper functions
def evaluate_model(name, model, X_train, y_train, X_test, y_test):
    """
    Trains a model and computes performance metrics on the test set.
    
    Returns:
        metrics: dictionary of evaluation metrics
        y_pred: predicted class labels
        y_proba: predicted probabilities for class '1'
    """
    # Train the model
    model.fit(X_train, y_train)
    
    # Predict class labels
    y_pred = model.predict(X_test)
    
    # Predict probability for positive class (digit '3')
    y_proba = model.predict_proba(X_test)[:, 1]
    
    # Compute evaluation metrics
    metrics = {
        "Model": name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred, zero_division=0),
        "Recall": recall_score(y_test, y_pred, zero_division=0),
        "F1-score": f1_score(y_test, y_pred, zero_division=0),
        "AUC": roc_auc_score(y_test, y_proba),
        "Average Precision": average_precision_score(y_test, y_proba)
    }
    
    return metrics, y_pred, y_proba


def plot_roc_pr_curves(y_test, proba_dict):
    """
    Plots ROC and Precision-Recall curves for multiple models.
    
    Args:
        y_test: true binary labels
        proba_dict: dictionary of model names -> predicted probabilities
    """
    plt.figure(figsize=(12, 5))

    # --- ROC Curves ---
    plt.subplot(1, 2, 1)
    for name, y_proba in proba_dict.items():
        fpr, tpr, _ = roc_curve(y_test, y_proba)
        auc = roc_auc_score(y_test, y_proba)
        plt.plot(fpr, tpr, lw=2, label=f"{name} (AUC={auc:.3f})")
    plt.plot([0, 1], [0, 1], 'k--', lw=1)
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curves (Trade-off between TPR and FPR)")
    plt.legend()

    # --- Precision-Recall Curves ---
    plt.subplot(1, 2, 2)
    for name, y_proba in proba_dict.items():
        precision, recall, _ = precision_recall_curve(y_test, y_proba)
        ap = average_precision_score(y_test, y_proba)
        plt.plot(recall, precision, lw=2, label=f"{name} (AP={ap:.3f})")
    plt.xlabel("Recall")
    plt.ylabel("Precision")
    plt.title("Precision‚ÄìRecall Curves (Positive class focus)")
    plt.legend()

    plt.tight_layout()
    plt.show()


# STEP 3: Define models

# 1Ô∏è Unregularized Logistic Regression (very weak L1 regularization)
logreg_unreg = LogisticRegression(
    penalty='l1', solver='liblinear', C=1e6, max_iter=1000
)

# 2Ô∏è Regularized Logistic Regression (best C from Section 5.2)
logreg_reg = LogisticRegression(
    penalty='l1', solver='liblinear', C=best_C, max_iter=1000
)

# 3Ô∏è Random Forest Classifier (default parameters)
rf_clf = RandomForestClassifier(random_state=42)


# STEP 4: Train and evaluate all models

metrics_list = []  # store metrics for each model
proba_dict = {}    # store predicted probabilities for ROC/PR plots

for name, model in [
    ("Unregularized Logistic Regression", logreg_unreg),
    (f"Regularized Logistic Regression (C={best_C})", logreg_reg),
    ("Random Forest", rf_clf)
]:
    metrics, y_pred, y_proba = evaluate_model(name, model, X_train, y_train, X_test, y_test)
    metrics_list.append(metrics)
    proba_dict[name] = y_proba
    print(f"‚úÖ Model trained and evaluated: {name}")


# STEP 5: Display metrics table

results_df = pd.DataFrame(metrics_list)

# Only format numeric columns to avoid ValueError with string 'Model' column
numeric_cols = results_df.select_dtypes(include=np.number).columns

display(
    results_df.style
    .background_gradient(cmap="Blues")
    .format({col: "{:.4f}" for col in numeric_cols})
)


# STEP 6: Plot ROC and Precision-Recall curves

plot_roc_pr_curves(y_test, proba_dict)



```

   ‚Ä¢ May overfit since MNIST images have 784 features.

2Ô∏è. Regularized Logistic Regression (C = best_C)
